```python
def read_dtrs(filenames, topfile):
    trajs = []
    for f in filenames:
        trajs.append(md.load_dtr(f, top=topfile))
    return md.join(trajs)
```

```python
def find_dtrs(d):
    dtrs = []
    jsdir = f"{d}/jobsteps"
    for js in os.listdir(jsdir):
        ts = f"{jsdir}/{js}/run.dtr/timekeys"
        if os.path.exists(ts):
            dtrs.append(f"{jsdir}/{js}/run.dtr")
    return dtrs
```
